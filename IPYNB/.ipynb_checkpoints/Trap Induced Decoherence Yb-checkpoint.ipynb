```python
# 定义函数计算态演化 #


def Solve_Ham(H,in_sta,tlist):

    t_sta = sesolve(H,in_sta,tlist)
    
    return t_sta
```

```python
# 构造哈密顿量和初态 #



# Gaussian Sampling #
# mag = para*(vector_x + para_ratio_list[para_num]*tensor_x) #
#loc = 0.433125*(0.8)  # 均值
loc = 0.433125*(0.8)  # 均值
#scale = 0.05  # 标准差
scale = 0  # 标准差
size = 100  # 抽样数
para_ratio_list = random.normal(loc,scale,size)  # 不同原子感受到的磁场

para = 1

# initial state #
in_state = basis(int(2*F+1),0)

```

```python
# decoherence - tau #


# precession #
tau_step = 1000  # precession step
tau_de_list = linspace(0.01*pi,10*pi,tau_step)  # precession time

abs_state = zeros((size,tau_step,int(2*F+1),1))

# density matrix #
rou = zeros((size,tau_step,int(2*F+1),int(2*F+1)))
rou16 = zeros((size,tau_step))

for para_num in range(size):
    
    mag = para*(2.72*vector_y + para_ratio_list[para_num]*tensor_x)
    #mag = para*(2.72*vector_y + 2.72*random.random()*tensor_x)
    
    # precession #
    in_state_de = in_state
    
    t_state = Solve_Ham(mag,in_state_de,tau_de_list)
    
    abs_state[para_num] = (absolute(t_state.states))**2
    
    for rou_num in range(tau_step):
        rou[para_num,rou_num] = t_state.states[rou_num]*(conjugate(transpose(t_state.states[rou_num])))  # density matrix
        rou16[para_num,rou_num] = rou[para_num,rou_num,0,int(2*F)]  # ρ(1,6)
    
```

```python
# decoherence - tau figure #

 
plt.figure(figsize=(20,8))

population = zeros((tau_step,int(2*F+1)))

for ele_num in range(int(2*F+1)):
    
    str1 = str(int(2*F)-ele_num*2)
    str2 = '/2'
    
    for tau_num in range(tau_step):
        
        population[tau_num,ele_num] = sum(abs_state[:,tau_num,ele_num])/size
        
    plt.plot(tau_de_list/pi,population[:,ele_num],label = str1+str2)
    plt.legend()
    
plt.xlabel('Time of Precession(π)')
plt.ylabel('Final Population')
plt.title('Trap Induced Decoherence')
plt.show

plt.figure(figsize=(20,8))
plt.plot(tau_de_list/pi,population[:,0],label='+5/2')  # mF=5/2
plt.plot(tau_de_list/pi,population[:,int(2*F)],label='- 5/2')  # mF=-5/2
plt.legend()
plt.xlabel('Time of Precession(π)')
plt.ylabel('Population')
plt.title('Trap Induced Decoherence')
plt.show

plt.figure(figsize=(20,8))
plt.plot(tau_de_list/pi,(population[:,0]-population[:,int(2*F)]))  # mF=5/2 - mF=-5/2
plt.xlabel('Time of Precession(π)')
plt.ylabel('Differential Signal')
plt.title('Trap Induced Decoherence')
plt.show

plt.figure(figsize=(20,8))
plt.plot(tau_de_list/pi,(population[:,0]-population[:,int(2*F)])/(population[:,0]+population[:,int(2*F)]))  # (mF=5/2 - mF=-5/2)/(mF=5/2 + mF=-5/2)
plt.xlabel('Time of Precession(π)')
plt.ylabel('Normalized Differential Signal')
plt.title('Trap Induced Decoherence')
plt.show

Rou16 = zeros(tau_step)
for tau_num in range(tau_step):
    
    Rou16[tau_num] = absolute(sum(rou16[:,tau_num]))/size  # Modulus of ρ(1,6)

plt.figure(figsize=(20,8))
plt.plot(tau_de_list/pi,Rou16)
plt.xlabel('Time of Precession(π)')
plt.ylabel('Modulus of ρ(1,6)')
plt.title('Trap Induced Decoherence')
plt.show
```
